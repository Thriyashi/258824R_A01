```python
#Question 01. (a) Gamma correction with γ=0.5

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

f = cv.imread('runway.png')
assert f is not None, "Image not found"

gamma = 0.5
t = np.array(255 * (f / 255) ** gamma, dtype=np.uint8)
plt.figure(figsize=(10, 5))
plt.imshow(t)
plt.title('Gamma Correction with γ=0.5')
plt.axis('off')
plt.show()
```

```python
#Contrast stretching

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

f = cv.imread('runway.png')
assert f is not None, "Image not found"

r1, r2 = 140, 200

# Apply piecewise transformation
t = np.zeros_like(f)
t[f <= r1] = 0
t[(f > r1) & (f < r2)] = ((f[(f > r1) & (f < r2)] - r1) / (r2 - r1)) * 255
t[f >= r2] = 255
plt.imshow(t)
plt.title('Contrast Stretching')
plt.axis('off')
plt.show()
```
