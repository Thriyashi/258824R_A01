```python
#Question 6 part (a)
#first–order partial derivatives
import numpy as np

#Define sigma
sigma = 2

#create grid
x,y = np.meshgrid(np.linspace(-5,5,100),np.linspace(-5,5,100))

#Gaussian function
G = (1/(2*np.pi*sigma**2))*np.exp(-(x**2+y**2)/(2*sigma**2))

#Numerical derivatives
Gx = np.gradient(G, axis=1) 
Gy = np.gradient(G, axis=0)

#Analytical derivatives
Gx_analytical = -(x/(sigma**2))*G
Gy_analytical = -(y/(sigma**2))*G

#compare difference
print("Max difference in Gx:", np.max(np.abs(Gx - Gx_analytical)))
print("Max difference in Gy:", np.max(np.abs(Gy - Gy_analytical)))
```
