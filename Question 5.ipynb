```python
#Question 5 (a)
#normalized 5×5 Gaussian kernel for σ = 2

import numpy as np
def gaussian_kernel(size, sigma):
    k = size // 2
    x, y = np.mgrid[-k:k+1, -k:k+1]
    kernel = np.exp(-(x**2 + y**2) / (2 * sigma**2))
    return kernel / np.sum(kernel) #normalize the kernel    

kernel = gaussian_kernel(5, 2)
print("5×5 Gaussian kernel for σ = 2:")
print(kernel) 
```

```python
#Question 05 part (b)
#3D visualization
import cv2 as cv
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

kernel = gaussian_kernel(51, 2)

x = np.arange(51)
y = np.arange(51)
X, Y = np.meshgrid(x, y)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, kernel, cmap='viridis')

ax.set_title('3D Visualization of 51×51 Gaussian Kernel (σ = 2)')
plt.show()

#Question 05 part (c)
#applying Gaussian smoothing to an image using the Gaussian Kernel

f = cv.imread('einstein.png')
assert f is not None, "Image not found. Please check the path."

smoothed_manual = cv.filter2D(f, -1, kernel)
plt.figure(figsize=(6, 4))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(f, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('Smoothed Image')
plt.imshow(smoothed_manual, cmap='gray')
plt.show()


#Question 4 (d)
#apply OpenCV GaussianBlur

smoothed_opencv = cv.GaussianBlur(f, (51, 51), 2)
plt.figure(figsize=(6, 4))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(f, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('Smoothed Image (GaussianBlur)')
plt.imshow(smoothed_manual, cmap='gray')
plt.figure(figsize=(6, 4))
plt.title('OpenCV GaussianBlur')
plt.imshow(smoothed_opencv, cmap='gray')
plt.show()


```
